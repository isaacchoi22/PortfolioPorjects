```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data

df = pd.read_csv(r'C:\Users\chlwl\Downloads\movies.csv')

#.head() a lot of the code to fit github
```

```python
# Looking at the data

df.head()
```

```python
# Seeing if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# dropping missing data

df = df.dropna()

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# changing data type of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# correct

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
df.sort_values(by=['gross'], inplace = False, ascending = False).head()
```

```python
# see all the data
pd.set_option('display.max_rows', None)
```

```python
# drop any duplicates

df.drop_duplicates().head()
```

```python
# searhing for any Outliers

df.boxplot(column=['gross'])
```

```python
# Scatter plot with budget vs gross


plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Buddget for Film')

plt.show
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x = "gross", y = "budget", data = df, scatter_kws = {'color' : 'red'}, line_kws = {'color' : 'blue'})
```

```python
df.corr(method = 'pearson') #pearson, kendall, spearman correlation 

# High correaltion between budget and gross
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df_numerized.corr(method='pearson')
corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
